```python
from Crypto.Hash import SHAKE128

# this is from sim_hash_stub matrix_script.py
def samplentt(seed:bytes) -> list:
    h = SHAKE128.new() # XOF.Init()
    h.update(seed) # XOF.Absorb
    q = 3329
    j = 0
    a = [0]*256 # output array
    while j < 256:
        temp = h.read(3) # XOF.Squeeze(3)
        d1 = temp[0] + (temp[1] & 0xf) << 8
        d2 = (temp[1] >> 4) + (temp[2] << 4)
        if d1 < q:
            a[j] = d1
            j = j + 1
        if (d2 < q) and (j < 256):
            a[j] = d2
            j = j + 1
    return a
def matrix_message_convert(m:str,salt1:int,salt2:int)->bytes:
    m = m+"%.2x%.2x"%(salt1,salt2)
    b = bytes.fromhex(m) # convert from hex string to bytes
    return b

def samplentt_suite(m:str,salt1:int,salt2:int):
    mm = matrix_message_convert(m,salt1,salt2)
    a = samplentt(mm)
    return a
    

m = "0708090a0b0c0d0e0f101112131415161718191a1b1c1d1e1f20212223242526"
# matrix hash require two salts
salt1 = 0
salt2 = 0
# mm
a = samplentt_suite(m,salt1,salt2)
# message_convert(m,salt1).decode("utf-8")
# mm = message_convert(message_convert(m,salt1).decode(),salt2)
print(a)

```
